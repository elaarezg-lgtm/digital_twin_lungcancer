```python
import pandas as pd

def inspect_excel_sheets(file_path):
    print(f"\n--- Inspection des onglets de : {os.path.basename(file_path)} ---")
    try:
        xl = pd.ExcelFile(file_path)
        sheets = xl.sheet_names
        print(f"Onglets trouvés : {sheets}")
        for sheet in sheets:
            print(f"\nAperçu de l'onglet : {sheet}")
            df = pd.read_excel(file_path, sheet_name=sheet, nrows=5)
            display(df.head())
    except Exception as e:
        print(f"Erreur lors de la lecture : {e}")

# Inspection des deux fichiers problématiques
inspect_excel_sheets(f"{path}/S_Table_13_RPPA.xlsx")
inspect_excel_sheets(f"{path}/S_Table_6_Copy_Number.xlsx")
```

```python
import pandas as pd
from google.colab import drive
import os

# 1. Montage du Drive
drive.mount('/content/drive')

# 2. Définition du chemin REEL (corrigé)
path = "/content/drive/MyDrive/pfe_digital_twin"

# 3. Chargement des fichiers (Utilisation de read_excel car ce sont des .xlsx)
try:
    # On charge le fichier clinique principal
    df_clinical = pd.read_excel(f"{path}/EGFR_Mutations_Table7_Cleaned.xlsx")

    # On charge le fichier RPPA (On utilise la table 13 pour l'instant)
    df_rppa = pd.read_excel(f"{path}/S_Table_13_RPPA.xlsx", header=1)

    # 4. Nettoyage des noms de colonnes
    df_clinical.columns = df_clinical.columns.str.strip()
    df_rppa.columns = df_rppa.columns.str.strip()

    # 5. Préparation de la fusion
    # Note: On utilise 'Tumor ID' qui est présent dans ton fichier clinique
    # Si df_rppa n'a pas de 'Tumor ID', la fusion retournera 0 lignes
    if 'Tumor ID' in df_rppa.columns:
        master_twin = pd.merge(df_clinical, df_rppa, on='Tumor ID', how='inner')
    else:
        print("⚠️ Attention: 'Tumor ID' absent de la table RPPA. Fusion impossible en l'état.")
        master_twin = df_clinical # On garde au moins les données cliniques

    # 6. Sauvegarde finale
    output_file = f"{path}/Master_Twin_Data_Final.csv"
    master_twin.to_csv(output_file, index=False)

    print(f"✅ Terminé ! Le fichier a été sauvegardé ici : {output_file}")
    print(f"Nombre de patients : {master_twin.shape[0]}")

except FileNotFoundError as e:
    print(f"❌ Erreur : Fichier introuvable. Vérifie le chemin : {e}")
```

```python
import pandas as pd
import os
from google.colab import drive

# 1. Montage
drive.mount('/content/drive')

# 2. Chemins
base_path = "/content/drive/MyDrive/pfe_digital_twin"

# 3. Chargement
df_rppa = pd.read_csv("/content/data_rppa.txt", sep='\t')
df_egfr = pd.read_excel(f"{base_path}/EGFR_Mutations_Table7_Cleaned.xlsx")

# 4. Transposition du RPPA
gene_col = df_rppa.columns[0]
df_rppa = df_rppa.set_index(gene_col).T.reset_index().rename(columns={'index': 'Tumor ID'})

# 5. Harmonisation des IDs (Crucial !)
# On transforme 'TCGA-05-4382-01' en 'TCGA-05-4382' pour correspondre au fichier clinique
df_egfr['Tumor ID'] = df_egfr['Tumor ID'].str.strip().str[:12]
df_rppa['Tumor ID'] = df_rppa['Tumor ID'].str.strip().str[:12]

# 6. Fusion
master = pd.merge(df_egfr, df_rppa, on='Tumor ID', how='inner')

# 7. Sauvegarde
output_path = f"{base_path}/Master_Twin_Data_Final.csv"
master.to_csv(output_path, index=False)

print(f"✅ Fusion réussie : {master.shape[0]} patients identifiés.")
print(f"Fichier sauvegardé : {output_path}")
```

```python
import pandas as pd

# 1. Chargement du fichier Master existant (Clinique + RPPA)
path = "/content/drive/MyDrive/pfe_digital_twin"
master_rppa = pd.read_csv(f"{path}/Master_Twin_Data_Final.csv")

# 2. Chargement et préparation des données CNA
df_cna_raw = pd.read_csv("/content/data_cna.txt", sep='\t')

# Transposition : les gènes (Hugo_Symbol) deviennent des colonnes, les patients des lignes
# On ignore la colonne 'Entrez_Gene_Id' pour la transposition
df_cna = df_cna_raw.drop(columns=['Entrez_Gene_Id']).set_index('Hugo_Symbol').T.reset_index()
df_cna = df_cna.rename(columns={'index': 'Tumor ID'})

# 3. Harmonisation des IDs à 12 caractères
master_rppa['Tumor ID'] = master_rppa['Tumor ID'].str.strip().str[:12]
df_cna['Tumor ID'] = df_cna['Tumor ID'].str.strip().str[:12]

# 4. Fusion finale (Multi-Omique)
# On utilise 'inner' pour garder uniquement les patients ayant toutes les données
master_multi = pd.merge(master_rppa, df_cna, on='Tumor ID', how='inner')

# 5. Sauvegarde
output_multi = f"{path}/Master_MultiOmics_27.csv"
master_multi.to_csv(output_multi, index=False)

print(f"✅ Fusion Multi-Omique réussie !")
print(f"Nombre final de patients : {master_multi.shape[0]}")
print(f"Nombre total de colonnes (Clinique + Protéines + CNA) : {master_multi.shape[1]}")
print(f"Fichier sauvegardé : {output_multi}")

display(master_multi.head())
```

```python
import pandas as pd
import os

# 1. Configuration des chemins
path = "/content/drive/MyDrive/pfe_digital_twin"
master_path = f"{path}/Master_Twin_Data_Final.csv"
cna_file = "/content/data_cna.txt"

# 2. Chargement des données
df_master = pd.read_csv(master_path)
df_cna_raw = pd.read_csv(cna_file, sep='\t')

# 3. Préparation du CNA (Transposition)
# On retire Entrez_Gene_Id, on met Hugo_Symbol en index, puis on transpose
df_cna = df_cna_raw.drop(columns=['Entrez_Gene_Id']).set_index('Hugo_Symbol').T.reset_index()
df_cna = df_cna.rename(columns={'index': 'Tumor ID'})

# 4. Harmonisation des IDs (Troncature à 12 caractères)
df_master['Tumor ID'] = df_master['Tumor ID'].str.strip().str[:12]
df_cna['Tumor ID'] = df_cna['Tumor ID'].str.strip().str[:12]

# 5. Fusion Multi-Omique (Clinique + RPPA + CNA)
master_multi = pd.merge(df_master, df_cna, on='Tumor ID', how='inner')

# 6. Sauvegarde du fichier final
output_multi = f"{path}/Master_MultiOmics_27.csv"
master_multi.to_csv(output_multi, index=False)

print(f"✅ Fusion Multi-Omique terminée avec succès !")
print(f"📊 Nombre de patients : {master_multi.shape[0]}")
print(f"🧬 Nombre total de variables (colonnes) : {master_multi.shape[1]}")
print(f"💾 Fichier sauvegardé : {output_multi}")

display(master_multi.head())
```
